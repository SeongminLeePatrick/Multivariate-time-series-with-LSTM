```python
energy = pd.read_pickle('./data/energy.pkl').drop(columns = ['Hour'])
energy.rename(columns = {'T':'temp'}, inplace = True)
energy
```

```python
import holidays
```

```python
holiday_list = []
for i in holidays.UnitedStates(years = np.arange(2006, 2014)).items():
    holiday_list.append(i[0].strftime("%Y-%m-%d"))
```

```python
energy['date'] = [i.strftime("%Y-%m-%d") for i in energy.index]
```

```python
energy_df = energy.groupby(['date'])['load'].sum().reset_index()
```

```python
energy_df['temp'] = energy.groupby(['date'])[ 'temp'].mean().reset_index()['temp']
```

```python
energy_df.index = pd.to_datetime(energy_df['date'])
```

```python
energy_df
```

```python
energy_df['holiday_flag'] = [1 if d in holiday_list else 0 for d in energy_df['date']]
```

```python
energy_df.drop(columns = ['date'], inplace = True)
```

```python
energy_df.to_pickle('./energy_df.pkl')
```

```python
energy_df
```

```python
valid_start_dt = '2011-01-01'
test_start_dt = '2013-01-01'

T = 60
HORIZON = 30
```

```python
train = energy_df.copy()[energy_df.index < valid_start_dt][['load', 'temp', 'holiday_flag']]
```

```python
train
```

```python
ex = TimeSeriesTensor(train, 'load', HORIZON, {'X':(range(-T+1, 1), ['load', 'temp', 'holiday_flag'])}, freq = 'D')
```

```python
from sklearn.preprocessing import MinMaxScaler

y_scaler = MinMaxScaler()
y_scaler.fit(train[['load']])

X_scaler = MinMaxScaler()
train[['load', 'temp','holiday_flag']] = X_scaler.fit_transform(train)
```

```python
train
```

```python
tensor_structure = {'X':(range(-T+1, 1), ['load', 'temp', 'holiday_flag'])}
```

```python
tensor_structure
```

```python
train_inputs = TimeSeriesTensor(train, 'load', HORIZON, {'X':(range(-T+1, 1), ['load', 'temp', 'holiday_flag'])}, freq = 'D')
```

```python
train_inputs2 = TimeSeriesTensor(train, 'holiday_flag', HORIZON, {'X':(range(-T+1, 1), ['load', 'temp', 'holiday_flag'])}, freq = 'D')
```

```python
train_inputs_decoder = train_inputs2['target']
```

```python
train_inputs_decoder = train_inputs_decoder[...,tf.newaxis]
```

```python
look_back_dt = dt.datetime.strptime(valid_start_dt, '%Y-%m-%d') - dt.timedelta(days=T-1)
valid = energy_df.copy()[(energy_df.index >=look_back_dt) & (energy_df.index < test_start_dt)][['load', 'temp', 'holiday_flag']]
valid[['load', 'temp', 'holiday_flag']] = X_scaler.fit_transform(valid)
valid_inputs = TimeSeriesTensor(valid, 'load', HORIZON, {'X':(range(-T+1, 1), ['load', 'temp', 'holiday_flag'])}, freq = 'D')
```

```python
valid_inputs = TimeSeriesTensor(valid, 'load', HORIZON, {'X':(range(-T+1, 1), ['load', 'temp', 'holiday_flag'])}, freq = 'D')
```

```python
valid_inputs2 = TimeSeriesTensor(valid, 'holiday_flag', HORIZON, {'X':(range(-T+1, 1), ['load', 'temp', 'holiday_flag'])}, freq = 'D')
```

```python
valid_input_decoder = valid_inputs2['target']
```

```python
valid_input_decoder = valid_input_decoder[...,tf.newaxis]
```

## tensor shape : (samples, time steps, features)
## input_shape = (timestep, features)

### T = 60(timestep), HORIZON = 30 (forecasting length)

```python
ENCODER_LAYER_1_DIM = 12
ENCODER_LAYER_2_DIM = 6
DECODER_DIM = 6
BATCH_SIZE = 32
EPOCHS = 15
```

```python
from tensorflow.keras.layers import Input, LSTM, Dense, RepeatVector, TimeDistributed, Flatten, concatenate
from tensorflow.keras.models import Model
```

```python
main_input = Input(shape = (T, train.shape[-1]), name = 'main_input') #input_shape = (timestep:60days, features: 3)
```

```python
encoder_lstm_first_out = LSTM(ENCODER_LAYER_2_DIM, return_sequences=True)(main_input)
```

```python
encoder_lstm_second_out = LSTM(ENCODER_LAYER_2_DIM)(encoder_lstm_first_out)
```

```python
decoder_out = RepeatVector(HORIZON)(encoder_lstm_second_out)
decoder_out2 = Dense(HORIZON)(decoder_out)
```

```python
aux_input = Input(shape=(30, 1), name = 'aux_input')
concat_decoder_out_aux_input = concatenate([decoder_out2, aux_input])
```

```python
decoder_lstm_out = LSTM(DECODER_DIM, return_sequences=True)(concat_decoder_out_aux_input)
time_based_out = TimeDistributed(Dense(1))(decoder_lstm_out)
flatten_out = Flatten()(time_based_out)
model = Model(inputs = [main_input, aux_input], outputs = [flatten_out])
```

```python
model.compile(optimizer='RMSprop', loss='mse')
```

```python
model.summary()
```

```python
EPOCHS = 150
```

```python
earlystop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=10)

history = model.fit([train_inputs['X'], train_inputs_decoder],train_inputs['target'],
              batch_size=BATCH_SIZE,
              epochs=EPOCHS,
              validation_data=([valid_inputs['X'], valid_input_decoder], valid_inputs['target']),
              callbacks=[earlystop])
```

```python
plot_df = pd.DataFrame.from_dict({'train_loss':history.history['loss'], 'val_loss':history.history['val_loss']})
plot_df.plot(logy=True, figsize=(10,10), fontsize=12)
plt.xlabel('epoch', fontsize=12)
plt.ylabel('loss', fontsize=12)
plt.show()
```
